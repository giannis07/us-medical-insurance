# U.S. Medical Insurance Costs

```python
#import csv library
import csv
```

```python
#create empty lists for every attriute
ages = []
sexes = []
bmis = []
num_of_children = []
smoker_statuses = []
regions = []
charges = []
```

```python
#load csv data
with open("insurance.csv",newline='') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        ages.append(row['age'])
        sexes.append(row['sex'])
        bmis.append(row['bmi'])
        num_of_children.append(row['children'])
        smoker_statuses.append(row['smoker'])
        regions.append(row['region'])
        charges.append(float(row['charges']))         
```

```python
# calculate the average for numerical values
def calculate_average(lst,lst_name):
    count = 0
    total= 0
    for i in range(len(lst)):
        total += float(lst[i])
        count += 1
    avg = total /count if count else 0
    print(f"Average {lst_name} is: {avg:.2f}")        
```

```python
calculate_average(ages,'age')
```

```python
calculate_average(bmis,'bmi')
```

```python
# calculate the average charge for categorical attributes (regions,sex,smoker)
def calculate_average_charge_by_category(lst,charges):
    # create dictionary for sums and counts for each category
    category_data ={}
    
    for i in range(len(lst)):
        category = lst[i]
        charge = charges[i]
        
        if category not in category_data:
            category_data[category] = {'total_charges':0,'count':0}
        category_data[category]['total_charges'] += charge
        category_data[category]['count'] += 1
    
    avg_charges = {}
    for category,data in category_data.items():
        if data['count'] > 0:
            avg_charges[category] = round(data['total_charges'] / data['count'],2)
        else:
            avg_charges[category] = 0 
    
    return avg_charges        
```

```python
avg_charges_by_region = calculate_average_charge_by_category(regions,charges)
print(f"Average charge by region:{avg_charges_by_region}")
```

```python
avg_charges_by_smoker = calculate_average_charge_by_category(smoker_statuses,charges)
print(f"Average charge by smoker status: {avg_charges_by_smoker}")
```

```python
avg_charges_by_sexes = calculate_average_charge_by_category(sexes,charges)
print(f"Average charge by sexes: {avg_charges_by_sexes}")
```

```python
# calculate the number of males and females in the file
def num_sexes():
    males = 0 
    females = 0
    
    for sex in sexes:
        if sex == 'male':
            males += 1
        elif sex == 'female':
            females += 1
    print(f"Number of males: {males}")
    print(f"Number of females: {females}")
```
